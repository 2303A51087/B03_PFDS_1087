```python
import pandas as pd
import numpy as np
days = pd.to_datetime(pd.Series(range(1, 31), name='Day'), unit='D', origin='2023-01-01').dt.strftime('%Y%m-%d')
temperatures = np.random.uniform(10, 30, size=30)
dummy_data = pd.DataFrame({'Day': days, 'Temperature': temperatures})
dummy_data.to_csv('daily_temperature.csv', index=False)
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("daily_temperature.csv")
plt.figure(figsize=(8,6))
plt.plot(data["Day"], data["Temperature"], marker="o", linestyle="-")
plt.xlabel("Day")
plt.ylabel("Temperature (C)")
plt.title("Daily Temperature Variation - Weather Dataset")
plt.grid(True)
plt.show()
hottest_day = data.loc[data["Temperature"].idxmax()]
coldest_day = data.loc[data["Temperature"].idxmin()]
print("Hottest Day:", hottest_day["Day"], "with", hottest_day["Temperature"], "C")
print("Coldest Day:", coldest_day["Day"], "with", coldest_day["Temperature"], "C")
```
