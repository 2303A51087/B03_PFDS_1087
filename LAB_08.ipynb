```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Load dataset
# df = pd.read_csv("population.csv") # Original line

# Load a sample dataset (replace with your actual data loading if available)
data = {'Country': ['China', 'India', 'United States', 'Indonesia', 'Pakistan'],
        'Population': [1444216107, 1393409038, 332915073, 276361783, 225199937]}
df = pd.DataFrame(data)


# Sort by Population and select top 5 countries
top5 = df.sort_values(by="Population", ascending=False).head(5)

# Custom formatter for y-axis (millions/billions)
def population_format(x, pos):
    if x >= 1_000_000_000:
        return f'{x/1_000_000_000:.1f}B'
    elif x >= 1_000_000:
        return f'{x/1_000_000:.1f}M'
    else:
        return f'{x}'

# Plot
plt.figure(figsize=(10,6))
bars = plt.bar(top5["Country"], top5["Population"], color="skyblue", edgecolor="black")

# Add values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, population_format(yval, None),
             ha='center', va='bottom', fontsize=10)

plt.title("Top 5 Countries by Population", fontsize=14, fontweight='bold')
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.gca().yaxis.set_major_formatter(FuncFormatter(population_format))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
# df = pd.read_csv("movies_genre.csv") # Original line

# Load a sample dataset (replace with your actual data loading if available)
data = {'Genre': ['Action', 'Comedy', 'Drama', 'Sci-Fi', 'Horror'],
        'Movie Count': [1200, 1500, 1800, 700, 900]}
df = pd.DataFrame(data)

# Sort by movie count (optional, for better visualization)
df = df.sort_values(by="Movie Count", ascending=False)

# Plot bar chart
plt.figure(figsize=(12,6))
bars = plt.bar(df["Genre"], df["Movie Count"], color="lightcoral", edgecolor="black")

# Add labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, str(yval),
             ha='center', va='bottom', fontsize=9)

plt.title("Number of Movies Produced by Genre", fontsize=14, fontweight="bold")
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
